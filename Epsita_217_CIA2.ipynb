Q1. Implement the Pre-processing steps in the corpus (given below).

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
```

```python
# Input corpus
corpus = "Transformer is an exceptional innovation in the field of Deep Learning, contributed by Ashish Vaswani et al. (2017), Google. The transformer is the most influential Neural Network model that has shown outstanding performance on various NLP tasks including Machine Reading Comprehension, Machine translation and sentence classification. Attention mechanism and parallelization are the prominent features in the transformers. Consequently, it can facilitate long-range dependencies without any gradient vanishing or gradient explosion problems and it overcomes the drawbacks of the existing methods such as RNN and LSTM. The transformer is executed with an encoder-decoder mechanism and the original article of transformers # “Attention All You Need”."
```

a. Word and Sentence Tokenization

```python
# Word Tokenization
words = word_tokenize(corpus)

# Sentence Tokenization
sentences = sent_tokenize(corpus)

print("Tokenized Words:")
print(words)

print("\nTokenized Sentences:")
print(sentences)
```

b. Stop words and
c. Punctation Removal

```python
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

# Remove punctuation
words_no_punct = [word for word in words if word not in string.punctuation]

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_no_punct if word.lower() not in stop_words]

print("Filtered Words:")
print(filtered_words)
```

d. Frequency Distribution

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Create frequency distribution
freq_dist = FreqDist(filtered_words)

# Visualize the frequency distribution
plt.figure(figsize=(12, 6))
freq_dist.plot(20, cumulative=False)
plt.show()
```

e. Stemming (use any 2 stemmers) and lemmatization

```python
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
nltk.download('wordnet')

# stemming and lemmatization
porter_stemmer = PorterStemmer()
snowball_stemmer = SnowballStemmer('english')
wordnet_lemmatizer = WordNetLemmatizer()
stemmed_words_porter = [porter_stemmer.stem(word) for word in filtered_words]
stemmed_words_snowball = [snowball_stemmer.stem(word) for word in filtered_words]
lemmatized_words = [wordnet_lemmatizer.lemmatize(word, pos="v") for word in filtered_words]

print("Stemmed Words (Porter):")
print(stemmed_words_porter)
print("\nStemmed Words (Snowball):")
print(stemmed_words_snowball)
print("\nLemmatized Words:")
print(lemmatized_words)
```

f. Tag the Words using PoS

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')

# Part of Speech (PoS) Tagging
pos_tags = pos_tag(filtered_words)

print("Words with PoS Tags:")
print(pos_tags)
```

g. Implement the Named Entity Recognition

```python
from nltk import ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')

words_pos_tags = [pos_tag(word_tokenize(sent)) for sent in sentences]

# Named Entity Recognition (NER)
ner_tagged_words = [ne_chunk(tagged_words) for tagged_words in words_pos_tags]
for ne_tagged_words in ner_tagged_words:
    print(ne_tagged_words)
```

Q2. Implement the Word2Vec

```python
from gensim.models import Word2Vec

text = "Mickey Mouse, a cheerful and optimistic mouse clad in red shorts and yellow shoes, is the iconic mascot of The Walt Disney Company. Debuting in 1928, this spunky character has charmed audiences for generations with his adventures and can-do attitude."

# Tokenize sentences and words
sentences = [word_tokenize(sentence) for sentence in sent_tokenize(text)]

# Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=0)

# similar words to 'Mickey'
similar_words = model.wv.most_similar('Mickey', topn=5)
print("Similar words to 'Mickey':", similar_words)
```

a. Visualize the vectors in a plane.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Extract vectors and corresponding words
vectors = [model.wv[word] for word in model.wv.key_to_index.keys()]
words = [word for word in model.wv.key_to_index.keys()]

# PCA
pca_model = PCA(n_components=2)
vectors_2d = pca_model.fit_transform(vectors)

# Plot the 2D vectors
plt.figure(figsize=(10, 7))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1])
for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]))
plt.show()
```

b. Write (type) the working and flow of the code.

Word2Vec is used for word embeddings.
It is a vector representations of words in a continuous vector space.
It is based on a shallow neural network model.
It learns to map words to high-dimensional vectors in such a way that words with similar meanings are closer together in the vector space.

There are two types of Word2Vec models:

1. Continuous Bag-of-Words (CBOW): It is used to predict the target word based on its context words within the surrounding words. The input to the model is a set of context words, and the output is the target word.
The model is trained to minimize the prediction error by adjusting the word vectors. It is effective for small datasets and frequent words.

2. Skip-gram: It is used to predict the context words (surrounding words) when a target word is given. The input to the model is a target word, and the output is a set of context words. It is efficient for large datasets and infrequent words.

Difference between CBOW and Skip-gram models are:

1. input-output relationships: CBOW predicts a word based on its context, while Skip-gram predicts the context when a word is given.

2. application: CBOW is effective for frequent words, while Skip-gram is more suitable for infrequent words.
